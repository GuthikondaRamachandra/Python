```
from pyspark import SparkConf, SparkContext
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit
from pyspark.sql.types import StructType,StringType,StructField
from xml.etree import ElementTree as ET
from pyspark.sql.functions import *
#read all xmls and build the data frame
def parse_xml(rdd):
    """
    Read the xml string from rdd, parse and extract the elements,
    then return a list of list.
    """
    results = []
    tree=ET.fromstring(rdd[0])
    seller=tree.findall("seller")
    loan_list=[]
    for p in seller[0]:
      if p.tag=="id":
        print(p.text)
        loan_list.append(p.text)
    trade = tree.findall("trade")
    for t in trade:
      print(t.attrib['id'])
      loan_list.append(t.attrib['id'])
      loan_list.append(t.attrib['type'])
      
    for t in trade[0]:
    
      if t.tag=="advance":
          for tt in t:
              print(tt.text)
              loan_list.append(tt.text)
      if t.tag=="discount":
          for dd in t:
              print(dd.text)
              loan_list.append(dd.text)
      if t.tag=="expected_payment_date":

              print(t.text)
              loan_list.append(t.text)
      if t.tag=="settlement_date":

              print(t.text)
              loan_list.append(t.text)
      if t.tag=="arrears":
          for aa in t:
              print(aa.text)
              loan_list.append(aa.text)
      if t.tag=="payment":
          for pp in t:
              print(pp.text)
              loan_list.append(pp.text)
      if t.tag=="price_grade":
          print(t.text)
          loan_list.append(t.text)
      if t.tag=="currency":
          print(t.text)
          loan_list.append(t.text)

      if t.tag=="face_value":
          for aa in t:
              print(aa.text)
              loan_list.append(aa.text)
      if t.tag=="outstanding_principal":
          for aa in t:
              print(aa.text)
              loan_list.append(aa.text)

      if t.tag=="crystalised_loss":
          for aa in t:
              print(aa.text)
              loan_list.append(aa.text)

      if t.tag=="gross_yield":
          for aa in t:
              print(aa.text)
              loan_list.append(aa.text)
      
    results.append(loan_list)
    return results
schema = StructType([ StructField("seller_id", StringType(), True),\
                      StructField("trade_id", StringType(), True),\
                      StructField("trade_type", StringType(), True),\
                      StructField("advance_date", StringType(), True),\
                      StructField("advance_amount", StringType(), True),\
                      StructField("advance_amount_gbp", StringType(), True),\
                      StructField("advance_percentage", StringType(), True),\
                      StructField("discount_percentage", StringType(), True),\
                      StructField("discount_on", StringType(), True),\
                      StructField("expected_payment_date", StringType(), True),\
                      StructField("settlement_date", StringType(), True),\
                      StructField("arrears_in_arrears", StringType(), True),\
                      StructField("arrears_in_arrears_on_date", StringType(), True),\
                      StructField("payment_State", StringType(), True),\
                      StructField("price_grade", StringType(), True),\
                      StructField("currency", StringType(), True),\
                      StructField("face_value_amount", StringType(), True),\
                      StructField("face_value_amount_gbp", StringType(), True),\
                      StructField("outstanding_principal_amount", StringType(), True),\
                      StructField("outstanding_principal_amount_gbp", StringType(), True),\
                      StructField("crystalised_loss_amount", StringType(), True),\
                      StructField("crystalised_loss_date", StringType(), True),\
                      StructField("gross_yield_annualised", StringType(), True)
                    ])
path = '/FileStore/tables/*.xml'
xml_files_rdd = spark.read.text(path, wholetext=True).rdd   
records_rdd = xml_files_rdd.flatMap(parse_xml)
loadBook_df = records_rdd.toDF(schema)
display(loadBook_df)
```

```
#feature  # 1  : expected duration
df_filtered=loadBook_df.filter(loadBook_df.trade_type=="Standard")
df_with_exp_duration=df_filtered.withColumn("expected_duration",datediff(to_date(substring("expected_payment_date",0,10),'yyyy-MM-dd'),to_date("advance_date",'yyyy-MM-dd')))
display(df_with_exp_duration)
```

```
#feature  # 2  : number of previously settled trades for the seller
# Consider only records where settlement date is before advance_date
df_with_exp_duration_filtered=df_with_exp_duration.filter(to_date(df_with_exp_duration.settlement_date,'yyyy-MM-dd') > to_date(df_with_exp_duration.advance_date,'yyyy-MM-dd'))
#df_with_exp_duration_filtered.show()
df_with_exp_duration_filtered=df_with_exp_duration_filtered.withColumnRenamed("settlement_date","settlement_date_filtered").withColumnRenamed("advance_date","advance_date_filtered").withColumnRenamed("seller_id","seller_id_filtered")
df_with_exp_duration_filtered_join=df_with_exp_duration.join(df_with_exp_duration_filtered,(df_with_exp_duration.seller_id==df_with_exp_duration_filtered.seller_id_filtered),"inner")
#display(df_with_exp_duration_filtered_join)
df_with_exp_duration_settl=df_with_exp_duration_filtered_join.filter(to_date(df_with_exp_duration_filtered_join.settlement_date,'yyyy-MM-dd') > to_date(df_with_exp_duration_filtered_join.settlement_date_filtered,'yyyy-MM-dd'))
#display(df_with_exp_duration_settl)
final_df=df_with_exp_duration_settl.groupBy(df_with_exp_duration_settl.settlement_date,df_with_exp_duration_settl.seller_id).agg(count(df_with_exp_duration_settl.seller_id).alias("previous_settled_trade_cont"))
#display(final_df)
final_loan_transformed_df=df_with_exp_duration.join(final_df,(df_with_exp_duration.seller_id==final_df.seller_id) & (df_with_exp_duration.settlement_date==final_df.settlement_date),"left").na.fill(0).drop(final_df.settlement_date).drop(final_df.seller_id)
display(final_loan_transformed_df)
```
